Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
p = ggplot(penguins, aes(x = body_mass_g))

p = p + geom_histogram(bins = 37)

p = p + facet_wrap(~year, nrow=3)

p = p + labs(y = "Number of Penguins", x = "Body Mass in Grams")

p = p + theme(panel.background = element_rect(fill = 'lightblue'))

p = p + theme(
    strip.text = element_text(hjust = 0.9, color = 'grey50', face = 'bold', size = 10),
    strip.background=element_blank())

p = p + scale_x_continuous(labels=scales::comma_format())

p
```

Is there a difference between the body mass of male and female penguins? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to compare within each species-group.**

```r
options(repr.plot.width = 4, repr.plot.height = 6)

p = ggplot(penguins, aes(x = body_mass_g, y = sex, color = sex)) 

p = p + geom_point(position = position_jitter(height = 0.25, width = 7), size = 1.2, color = 'gray40')

p = p + labs(x = "Body mass in grams", y = " ")

p = p + facet_wrap(~species, nrow = 3)

p = p + theme(panel.background = element_blank(),
     panel.grid.major = element_line(color = "gray90"))

p = p + theme(
    strip.text = element_text(hjust = 0.1, color = 'grey50', face = 'bold', size = 10),
    strip.background=element_blank())

p = p + scale_x_continuous(labels=scales::comma_format())

p
```
